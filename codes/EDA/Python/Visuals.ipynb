```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_text.csv")
df.head()
```

```python
from wordcloud import WordCloud as wc

school = df.loc[df["1"] == "school",]
education = df.loc[df["1"] == "education",]

words = []
for i in range(len(school)):
    for j in str(school.loc[i,"0"]).split(" "):
        if len(j) > 1:
            words.append(j)

word_cloud = wc(collocations = False, background_color = "white").generate(" ".join(words))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
```

```python
education = education.reset_index(drop=True)

words = []
for i in range(len(education)):
    for j in str(education.loc[i,"0"]).split(" "):
        if len(j) > 1:
            words.append(j)

word_cloud = wc(collocations = False, background_color = "white").generate(" ".join(words))
plt.imshow(word_cloud, interpolation="bilinear")
plt.axis("off")
```
